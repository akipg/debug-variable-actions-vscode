```python

def clip_text(text):
    import pyperclip
    import urllib.parse

    encoded_text = urllib.parse.quote(text)
    print(encoded_text)
    pyperclip.copy(encoded_text)


def encode_current_clip():
    import pyperclip
    import urllib.parse
    # Get the text from the clipboard
    clipboard_text = pyperclip.paste()

    encoded_text = urllib.parse.quote(clipboard_text)
    print(encoded_text)
    pyperclip.copy(encoded_text)
```

```python
def show_current_clip():
    import pyperclip
    import urllib.parse
    import xml.dom.minidom

    # Get the text from the clipboard
    clipboard_text = pyperclip.paste()

    # Decode the HTML
    decoded_text = urllib.parse.unquote(clipboard_text)

    try:
        # Pretty print the decoded HTML
        pretty_html = xml.dom.minidom.parseString(decoded_text).toprettyxml()

        print(pretty_html)
    except:
        # If there was an error, print the decoded text
        print(decoded_text)
        pass
show_current_clip()
```
